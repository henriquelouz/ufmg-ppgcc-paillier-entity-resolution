```python
df = pd.read_csv("tb_dados_pacientes40.csv", sep=';')
df["sexo"] = df["sexo"].apply(lambda x: 'M' if x == 1 else 'F')
df['cpf'] = df['cpf'].astype('Int64').astype('str').replace({'<NA>': ""})
df['cns'] = df['cns'].astype('Int64').astype('str').replace({'<NA>': ""})
df['cep'] = df['cep'].astype('Int64').astype('str').replace({'<NA>': ""})
#df = df.fillna('')
df.replace('', np.nan, inplace=True)
df = df.dropna()
df
```

```python
df_10 = pd.read_csv("df_10.csv").set_index("index")
df_10.index.name = None
df_10
```

```python
for COLUNA1 in df.columns.tolist():
    gps = df.groupby([COLUNA1]).size().reset_index(name='counts').sort_values('counts', ascending=False)
    df_blocking = df.drop(df.index)
    for i, row in gps.iterrows():
        #dfsilva = df.loc[(df[COLUNA1] == row[COLUNA1]) & (df[COLUNA2] == row[COLUNA2])]
        dfsilva = df.loc[(df[COLUNA1] == row[COLUNA1])]
        mcnomep = dfsilva.groupby(['nomep']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
        mcnomem = dfsilva.groupby(['nomem']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
        mcnomeu = dfsilva.groupby(['nomeu']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
        mcnomemaep = dfsilva.groupby(['nomemaep']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
        mcnomemaem = dfsilva.groupby(['nomemaem']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
        mcsexo = dfsilva.groupby(['sexo']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
        mcdata = dfsilva.groupby(['data']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
        mccpf = dfsilva.groupby(['cpf']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
        mccns = dfsilva.groupby(['cns']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
        mccodmunicipio = dfsilva.groupby(['codmunicipio']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
        mccep = dfsilva.groupby(['cep']).size().reset_index(name='counts').sort_values('counts', ascending=False).head(1).reset_index()
    
        data = {
            'nomep': [mcnomep.at[0, 'nomep']],
            'nomem': [mcnomem.at[0, 'nomem']],
            'nomeu': [mcnomeu.at[0, 'nomeu']],
            'nomemaep': [mcnomemaep.at[0, 'nomemaep']],
            'nomemaem': [mcnomemaem.at[0, 'nomemaem']],
            'sexo': [mcsexo.at[0, 'sexo']],
            'data': [mcdata.at[0, 'data']],
            'cpf': [mccpf.at[0, 'cpf']],
            'cns': [mccns.at[0, 'cns']],
            'codmunicipio': [mccodmunicipio.at[0, 'codmunicipio']],
            'cep': [mccep.at[0, 'cep']],
        }
    
        d = pd.DataFrame.from_dict(data)
        df_blocking = pd.concat([df_blocking, d])
    
    df_blocking = df_blocking.reset_index()
    
    correctly_placed = 0
    for a, row_10 in df_10.iterrows():
        df_blocking_match = df_blocking.copy(deep=True)
        df_blocking_match['match_score'] = 0
        
        for i, row in df_blocking_match.iterrows():
            for c, item in row.items():
                if c == "index" or c == "match_score":
                    continue
                    
                if item == df_10.at[a, c]:
                    df_blocking_match.at[i, 'match_score'] += 1
        
        df_blocking_match = df_blocking_match.sort_values('match_score', ascending=False)
        df_max = df_blocking_match[df_blocking_match['match_score'] == df_blocking_match['match_score'].max()]
        #if df_10.at[a, COLUNA1] in df_max[COLUNA1].to_list() and df_10.at[a, COLUNA2] in df_max[COLUNA2].to_list():
        if df_10.at[a, COLUNA1] in df_max[COLUNA1].to_list():
            correctly_placed+=1
            
    df_max_total = df_blocking.drop(df_blocking.index).drop("index", axis=1)
    
    for a, row_10 in df_10.iterrows():
        df_blocking_match = df_blocking.copy(deep=True)
        df_blocking_match['match_score'] = 0
        
        for i, row in df_blocking_match.iterrows():
            for c, item in row.items():
                if c == "index" or c == "match_score":
                    continue
                    
                if item == df_10.at[a, c]:
                    df_blocking_match.at[i, 'match_score'] += 1
        
        df_blocking_match = df_blocking_match.sort_values('match_score', ascending=False)
        df_max = df_blocking_match[df_blocking_match['match_score'] == df_blocking_match['match_score'].max()]
        df_max_total = pd.concat([df_max_total, df_max])
        
    total_comparsions = 0
    for i, row in df_max_total.iterrows():
        #total_comparsions += len(df[(df[COLUNA1] == row[COLUNA1]) & (df[COLUNA2] == row[COLUNA2])])
        total_comparsions += len(df[(df[COLUNA1] == row[COLUNA1])])
        
    print(30*"=")
    print(COLUNA1)
    blocos = len(gps)
    print(f"blocos: {blocos}")
    print(f'r/b: {gps.loc[:, 'counts'].mean():.7n}')
    
    db = correctly_placed
    de = 32
    e = len(df) * de
    print(f"e: {e}")
    
    r = blocos * de
    sumgp = total_comparsions
    b = r + sumgp
    
    CPb = db/de
    QPb = db / b
    TRbe = 1 - b / e
    
    print(f'CP(B) = {CPb*100:.7n}%')
    print(f'QP(B) = {QPb*100:.7n}%')
    print(f'TR(B,E) = {TRbe*100:.7n}%')
```
